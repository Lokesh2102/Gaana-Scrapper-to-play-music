```python
from selenium import webdriver
import requests
import bs4
```

```python
browser=webdriver.Chrome(r"C:\Users\user\Downloads\chromedriver")
browser.set_window_size(760,850)
```

```python
#Create selenium browser
url="https://gaana.com/search/"
print("Enter a song")
nam=input()
n=""
if(len(nam.split(" "))>1):
    for i in range(len(nam.split(" "))):
        n=n+nam.split(" ")[i]
        if(i<len(nam.split(" "))-1):
            n+="%20"
else:
    n=nam
url+=n
browser.get(url)
url_values=[]
song_title=[]
song_values=[]
content = requests.get(url)
soup = bs4.BeautifulSoup(content.text, 'html.parser')
te1=soup.find_all("div",class_="card_layout")
for i in te1:
    te2=i.find_all("div",class_="hover-events-parent")
    for j in te2:
        te3=j.find_all("a")
        url_values.append(te3[0].get('href'))
        song_title.append(te3[0].img.get('title'))
    break  

for i in te1:
    te2=i.find_all("div",class_="arwtork_label")
    for j in te2:
        te3=j.find("span",class_="desc_sl")
        song_values.append(te3.get_text())
    break
```

```python
count=1
for i in range(len(song_title)):
    print(count,end=" - ")
    print(song_title[i]+" ["+song_values[i]+"] ")
    count+=1
```

```python
url_o="https://gaana.com"
if(len(url_values)==1):
    url_o+=url_values[0]
else:
    print("Enter the number ---")
    num=int(input())
    url_o+=url_values[num-1]
```

```python
browser.get(url_o)
```

```python
content = requests.get(url_o)
soup = bs4.BeautifulSoup(content.text, 'html.parser')
try:
    te1=soup.find_all("div",class_="lyricstxt")[0].find_all("p")
    print("Lyrics for the song. Sing Along ---- :)")
    for i in te1:
        print(i.get_text())
except:
    print("No lyrics for the music")
```
